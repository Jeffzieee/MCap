```python
#Dependancies
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline #Merging different modules
from sklearn.metrics import r2_score #r_2 metrics
%matplotlib inline
```

```python
boston=load_boston()
print(boston.DESCR)
```

```python
#Accessing Data Attribute
for name, index in enumerate(boston.feature_names):
    print(index,name)
```

```python
#reshaping data
dataset = boston.data
data=dataset[:,12].reshape(-1,1)
np.shape(data)
```

```python
#reshaping target 
target = boston.target.reshape(-1,1)
np.shape(target)
```

```python
#checking the working of matplotlib 
plt.scatter(data,target,color='green')
plt.xlabel('Lower income population')
plt.ylabel('Cost of House')
plt.show()
```

```python
#using regression
reg=LinearRegression()
reg.fit(data,target)
```

```python
#Prediction
pred=reg.predict(data)
print(pred)
```

```python
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel('Lower income population')
plt.ylabel('Cost of House')
plt.show()
```

```python
model=make_pipeline(PolynomialFeatures(3), reg)
model.fit(data,target)
```

```python
predi=model.predict(data)
plt.scatter(data,target,color='green')
plt.plot(data,predi,color='red')
plt.xlabel('Lower income population')
plt.ylabel('Cost of House')
plt.show()
```

from sklearn.metrics import r2_score

```python
#r2(square)_metrics
from sklearn.metrics import r2_score
r2_score(predi,target)
```
